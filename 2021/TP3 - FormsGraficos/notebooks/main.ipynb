```python
import pandas as pd
import matplotlib.pyplot as gf1
import matplotlib.pyplot as gf2
import matplotlib.pyplot as gf3

data = pd.read_csv("../data/EncuestaForms.csv")
```

```python
x = data['Genero_musical_que_mas_gusta'].unique()
y = data['Genero_musical_que_mas_gusta'].value_counts().tolist()

gf1.bar(x,y)
gf1.title('Genero musical que mas gusta')
gf1.xlabel('Genero')
gf1.ylabel('Cantidad de personas que le gusta ese genero')
gf1.xticks(rotation=90,fontsize=12)
gf1.show()
```

```python
xx = data['Artista_o_Banda_que_mas_gusta'].value_counts().index.tolist()
yy = data['Artista_o_Banda_que_mas_gusta'].value_counts().tolist()

gf2.bar(xx,yy)
gf2.title('Artista o banda que mas gusta')
gf2.xlabel('Artista o banda')
gf2.ylabel('Cantidad de personas que le gusta ese artista o banda')
gf2.xticks(rotation=90,fontsize=12)
gf2.show()
```

```python
x3 = data['Nicki_Nicole_del_1_al_10'].unique()
y3 = data['Nicki_Nicole_del_1_al_10'].value_counts().tolist()

gf3.bar(x3,y3)
gf3.title('Cuanto gusta Nicki Nicole')
gf3.xlabel('Puntaje')
gf3.ylabel('Cantidad de personas que voto esa opcion')
gf3.xticks(rotation=90,fontsize=12)
gf3.show()
```
