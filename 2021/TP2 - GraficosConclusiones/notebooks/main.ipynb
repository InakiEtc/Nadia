```python
import pandas as pd
import matplotlib.pyplot as gf1
import matplotlib.pyplot as gf2
import numpy as np

data = pd.read_csv("../data/llamados-atendidos-violencia-familiar-202105.csv")
data = data.loc[0:100]
```

```python
x = data['violencia_tipo'].unique()
y = data['violencia_tipo'].value_counts().tolist()

gf1.bar(x,y)
gf1.title('Tipos de violencia')
gf1.xlabel('Tipos de Violencia')
gf1.ylabel('Cantidad')
gf1.xticks(rotation=90,fontsize=11)
gf1.show()
```

```python
xx = data['agresor_relacion_victima'].unique()
yy = data['agresor_relacion_victima'].value_counts().tolist()

gf2.plot(xx, yy)
gf2.title('Relacion Victima-Agresor')
gf2.xlabel('Relacion')
gf2.ylabel('Cantidad')
gf2.xticks(rotation=90,fontsize=11)
gf2.show()
```
