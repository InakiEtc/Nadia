```python
import pandas as pd
import seaborn as sb

data = pd.read_csv("../data/ellas-hacen-titulares-activos.csv",sep=",",engine="python")
grafico = sb.pairplot(data,hue="provincia",palette="icefire")
```

Como se puede ver en los graficos la provincia de Buenos Aires y Tucuman
son las que mas cantidad de titualres activos tienen en el progrma "Ellas hacen"
con mas de 6000 titualares repartidos en distintos departamentos.

En cambio la provincia de La rioja es la que menos titualres posee
